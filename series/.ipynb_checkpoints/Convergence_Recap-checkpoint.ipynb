# Understanding Convergence in Mathematics



Understanding the concept of **convergence** involves recognizing how different mathematical objects (such as sequences, series, integrals, and functions) approach specific values under certain conditions. Below, is an explanation for each of the listed concepts with definitions and examples, to clarify their meaning.

- **Improper Integral Convergence:** The integral evaluates to a finite number despite infinite intervals or discontinuities.

- **Sequence Convergence:** The terms approach a specific value as the index $ n $ increases indefinitely.

- **Series Convergence:** The sum of a sequence's terms approaches a finite limit.

- **Radius of Convergence for Power Series:** The range of $ x $ values for which the power series converges.

- **Function Convergence:** It's more accurate to discuss the convergence of a function's limit or an integral involving the function, rather than the function itself.

---

## 1. What Does Convergence of an Improper Integral Mean?



An **improper integral** is an integral where either the interval of integration is infinite, or the integrand becomes infinite within the interval. The **convergence** of an improper integral means that the integral evaluates to a finite number.



### Examples:



**a) Integral Over an Infinite Interval:**



$$\int_{1}^{\infty} \frac{1}{x^2} \, dx$$



**Solution:**



Compute the limit as the upper bound approaches infinity:


\begin{align*}
\int_{1}^{\infty} \frac{1}{x^2} \, dx &= \lim_{b \to \infty} \int_{1}^{b} \frac{1}{x^2} \, dx \\
&= \lim_{b \to \infty} \left( -\frac{1}{x} \Big|_{1}^{b} \right) \\
&= \lim_{b \to \infty} \left( -\frac{1}{b} + 1 \right) \\
&= 1
\end{align*}


Since the limit exists and is finite, the improper integral **converges** to 1.



---



**b) Integral with an Infinite Discontinuity:**



$$\int_{0}^{1} \frac{1}{\sqrt{x}} \, dx$$



**Solution:**



Address the discontinuity at $ x = 0 $:


\begin{align*}
\int_{0}^{1} \frac{1}{\sqrt{x}} \, dx &= \lim_{a \to 0^+} \int_{a}^{1} \frac{1}{\sqrt{x}} \, dx \\
&= \lim_{a \to 0^+} \left( 2\sqrt{x} \Big|_{a}^{1} \right) \\
&= \lim_{a \to 0^+} \left( 2(1) - 2\sqrt{a} \right) \\
&= 2
\end{align*}


The integral **converges** to 2.

---


**c) Non-Convergent Example:**

$$\int_{1}^{\infty} \frac{1}{x} \, dx$$

**Solution:**

\begin{align*}
\int_{1}^{\infty} \frac{1}{x} \, dx &= \lim_{b \to \infty} \int_{1}^{b} \frac{1}{x} \, dx \\
&= \lim_{b \to \infty} \left( \ln x \Big|_{1}^{b} \right) \\
&= \lim_{b \to \infty} \left( \ln b - \ln 1 \right) \\
&= \infty
\end{align*}

Since the limit is infinite, the improper integral **diverges**.



---

## 2. What Does Convergence of a Sequence Mean?

A **sequence** is an ordered list of numbers $ \{a_n\} $. A sequence **converges** to a limit $ L $ if its terms approach $ L $ as $ n $ approaches infinity. Formally, for every $ \varepsilon > 0 $, there exists an $ N $ such that $ |a_n - L| < \varepsilon $ for all $ n \geq N $.

### Examples:

**a) Convergent Sequence:**

Let $ a_n = \frac{1}{n} $.

As $ n \to \infty $, $ a_n \to 0 $.

**Explanation:**

For any $ \varepsilon > 0 $, choose $ N > \frac{1}{\varepsilon} $. Then for all $ n \geq N $:

$$\left| \frac{1}{n} - 0 \right| = \frac{1}{n} < \varepsilon$$

Thus, the sequence **converges** to 0.

---

**b) Divergent Sequence:**

Let $ a_n = (-1)^n $.

**Explanation:**

The sequence oscillates between 1 and -1 and does not approach any single value. Therefore, it **diverges**.

---

## 3. What Does Convergence of a Series Mean?



A **series** is the sum of the terms of a sequence:



$$S = \sum_{n=1}^{\infty} a_n$$



The series **converges** if the sequence of its partial sums $ S_N = \sum_{n=1}^{N} a_n $ approaches a finite limit as $ N \to \infty $.



### Examples:



**a) Convergent Series (Geometric Series):**



$$\sum_{n=0}^{\infty} \left( \frac{1}{2} \right)^n = \frac{1}{1 - \frac{1}{2}} = 2$$



**Explanation:**



Since the common ratio $ r = \frac{1}{2} $ satisfies $ |r| < 1 $, the series **converges** to 2.



---



**b) Divergent Series (Harmonic Series):**



$$\sum_{n=1}^{\infty} \frac{1}{n} = \infty$$



**Explanation:**



The partial sums grow without bound, so the series **diverges**.



---

## 4. Why Does a Power Series Have a Radius of Convergence, but a (Normal) Series Doesn't?



A **power series** is a series in the form:



$$\sum_{n=0}^{\infty} a_n (x - c)^n$$



- The variable $ x $ makes the convergence dependent on its value.

- The **radius of convergence** $ R $ is the distance from the center $ c $ within which the series converges.



A **normal series** $ \sum_{n=1}^{\infty} a_n $ lacks the variable $ x $ and thus doesn't have a radius of convergence; it either converges or diverges based on its terms alone.



### Examples:



**a) Power Series with Finite Radius:**



$$\sum_{n=0}^{\infty} x^n$$



**Solution:**



Using the ratio test:



\begin{align*}
L &= \lim_{n \to \infty} \left| \frac{x^{n+1}}{x^n} \right| = |x| \\
\end{align*}



- **Converges** if \( |x| < 1 \).

- **Diverges** if \( |x| > 1 \).



Thus, the radius of convergence $ R = 1 $.

---



**b) Power Series with Infinite Radius:**



$$\sum_{n=0}^{\infty} \frac{x^n}{n!}$$



**Solution:**



Using the ratio test:



\begin{align*}
L &= \lim_{n \to \infty} \left| \frac{x^{n+1}/(n+1)!}{x^n/n!} \right| \\
&= \lim_{n \to \infty} \left| \frac{x}{n+1} \right| = 0
\end{align*}

Since $ L = 0 $ for all $ x $, the radius of convergence $ R = \infty $.

---


## 5. Does It Make Sense to Say That a Function Converges or Diverges?



No, it doesn't make sense to say that a **function** converges or diverges in isolation. However, we can discuss:



- The **limit of a function** as the variable approaches a certain value.

- The **convergence of an improper integral** involving the function.



### Examples:



**a) Limit of a Function:**



Let $ f(x) = \frac{1}{x} $.



As $ x \to \infty $:



$$\lim_{x \to \infty} \frac{1}{x} = 0$$



The limit **converges** to 0.



---



**b) Function Without a Limit:**



Let $ f(x) = \sin x $.



As $ x \to \infty $:



$$\lim_{x \to \infty} \sin x$$



This limit does not exist because $ \sin x $ oscillates indefinitely between -1 and 1. Thus, the limit **diverges**.



---



**c) Improper Integral of a Function:**



$$\int_{1}^{\infty} \frac{\sin x}{x} \, dx$$



**Explanation:**



This integral **converges** due to the integrand's diminishing amplitude, despite the oscillation.


**Additional Note(if interested):**

This integral is known as the **sine integral** and is related to the **Dirichlet integral**. It converges but does not have an elementary closed-form solution. Instead, its value is known through numerical approximation:

$$\int_{1}^{\infty} \frac{\sin x}{x} \, dx \approx 0.504$$

Alternatively, this result can be expressed in terms of the **Sine Integral function** $ \text{Si}(x) $, a special function used to evaluate integrals of this form. Specifically:

$$\int_{1}^{\infty} \frac{\sin x}{x} \, dx = \text{Si}(\infty) - \text{Si}(1) \approx \frac{\pi}{2} - \text{Si}(1)$$

where $ \text{Si}(1) \approx 0.946 $, yielding an approximate value of $ 0.504 $ for the integral.
