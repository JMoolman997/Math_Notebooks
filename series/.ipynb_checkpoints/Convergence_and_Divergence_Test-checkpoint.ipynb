## Introduction

An **infinite series** is the sum of the terms of an infinite sequence. Determining whether such a series **converges** (approaches a finite limit) or **diverges** (does not approach a finite limit) is a fundamental question in calculus and analysis. Below are various tests and criteria used to analyze the convergence or divergence of series.

## Basic Definitions

- **Infinite Series**: $\sum_{n=1}^{\infty} a_n = a_1 + a_2 + a_3 + \cdots$
- **Partial Sum**: $S_N = \sum_{n=1}^{N} a_n$
- **Convergent Series**: A series $\sum_{n=1}^{\infty} a_n$ converges to $L$ if $\lim_{N \to \infty} S_N = L$.
- **Divergent Series**: A series that does not converge to any finite limit.

## Convergence Tests

### Nth-Term Test for Divergence

**Statement**: If $\lim_{n \to \infty} a_n \neq 0$, then the series $\sum_{n=1}^{\infty} a_n$ **diverges**.

**Formula**:

$$\lim_{n \to \infty} a_n \neq 0 \implies \sum_{n=1}^{\infty} a_n \text{ diverges}$$

### Geometric Series Test

**Statement**: A geometric series $\sum_{n=0}^{\infty} ar^n$ converges if $|r| < 1$ and diverges otherwise.

**Formula**:
$$
\sum_{n=0}^{\infty} ar^n =
\begin{cases}
\frac{a}{1 - r} & \text{if } |r| < 1, \\
\text{Diverges} & \text{if } |r| \geq 1.
\end{cases}
$$

### p-Series Test

**Statement**: The p-series $\sum_{n=1}^{\infty} \frac{1}{n^p}$ converges if $p > 1$ and diverges if $p \leq 1$.

**Formula**:
$$
\sum_{n=1}^{\infty} \frac{1}{n^p} =
\begin{cases}
\text{Converges} & \text{if } p > 1, \\
\text{Diverges} & \text{if } p \leq 1.
\end{cases}
$$

### Integral Test

**Statement**: Let $f(n) = a_n$ where $a_n = f(n)$, and $f$ is positive, continuous, and decreasing for $n \geq N$.

Then,
$
\sum_{n=N}^{\infty} a_n \text{ and } \int_{N}^{\infty} f(x) \, dx
$
either both converge or both diverge.

**Formula**:
$$
\sum_{n=N}^{\infty} a_n \text{ converges} \iff \int_{N}^{\infty} f(x) \, dx \text{ converges}
$$

### Comparison Test

#### Direct Comparison Test

**Statement**: Let $0 \leq a_n \leq b_n$ for all $n \geq N$.

- If $\sum b_n$ **converges**, then $\sum a_n$ **converges**.
- If $\sum a_n$ **diverges**, then $\sum b_n$ **diverges**.

**Formula**:
$$
0 \leq a_n \leq b_n \quad \forall n \geq N
$$
$$
\sum b_n \text{ converges} \implies \sum a_n \text{ converges}
$$
$$
\sum a_n \text{ diverges} \implies \sum b_n \text{ diverges}
$$

#### Limit Comparison Test

**Statement**: Let $a_n > 0$ and $b_n > 0$ for all $n$. If
$
\lim_{n \to \infty} \frac{a_n}{b_n} = c
$
where $c$ is a positive finite number, then $\sum a_n$ and $\sum b_n$ either both **converge** or both **diverge**.

**Formula**:
$$
\lim_{n \to \infty} \frac{a_n}{b_n} = c \in (0, \infty) \implies \sum a_n \text{ and } \sum b_n \text{ have the same behavior}
$$

### Alternating Series Test (Leibniz's Test)

**Statement**: Consider an alternating series $\sum_{n=1}^{\infty} (-1)^{n-1} b_n$ where $b_n > 0$.

If
1. $\lim_{n \to \infty} b_n = 0$, and
2. $b_{n+1} \leq b_n$ for all $n$ sufficiently large,

then the series **converges**.

**Formula**:
$$
\sum_{n=1}^{\infty} (-1)^{n-1} b_n \text{ converges if } \lim_{n \to \infty} b_n = 0 \text{ and } b_n \text{ is decreasing}
$$

### Absolute Convergence Test

**Statement**: If $\sum |a_n|$ **converges**, then $\sum a_n$ **converges absolutely**. Absolute convergence implies convergence, but not vice versa.

**Formula**:
$$
\text{If } \sum |a_n| \text{ converges, then } \sum a_n \text{ converges absolutely}
$$

### Ratio Test

**Statement**: Let $\rho = \lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right|$.

- If $\rho < 1$, the series $\sum a_n$ **converges absolutely**.
- If $\rho > 1$, the series $\sum a_n$ **diverges**.
- If $\rho = 1$, the test is **inconclusive**.

**Formula**:
$$
\rho = \lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right|
$$
$$
\begin{cases}
\rho < 1 & \implies \text{Converges Absolutely}, \\
\rho > 1 & \implies \text{Diverges}, \\
\rho = 1 & \implies \text{Inconclusive}.
\end{cases}
$$

### Root Test

**Statement**: Let $\rho = \lim_{n \to \infty} \sqrt[n]{|a_n|}$.

- If $\rho < 1$, the series $\sum a_n$ **converges absolutely**.
- If $\rho > 1$, the series $\sum a_n$ **diverges**.
- If $\rho = 1$, the test is **inconclusive**.

**Formula**:
$$
\rho = \lim_{n \to \infty} \sqrt[n]{|a_n|}
$$
$$
\begin{cases}
\rho < 1 & \implies \text{Converges Absolutely}, \\
\rho > 1 & \implies \text{Diverges}, \\
\rho = 1 & \implies \text{Inconclusive}.
\end{cases}
$$

### Cauchy's Condensation Test

**Statement**: Let $\{a_n\}$ be a non-increasing sequence of positive real numbers. Then, the series $\sum_{n=1}^{\infty} a_n$ **converges** if and only if the condensed series $\sum_{k=0}^{\infty} 2^k a_{2^k}$ **converges**.

**Formula**:
$$
\sum_{n=1}^{\infty} a_n \text{ converges} \iff \sum_{k=0}^{\infty} 2^k a_{2^k} \text{ converges}
$$

### Raabe's Test

**Statement**: Let $\rho = \lim_{n \to \infty} n \left( \frac{a_n}{a_{n+1}} - 1 \right)$.

- If $\rho > 1$, the series $\sum a_n$ **converges**.
- If $\rho < 1$, the series $\sum a_n$ **diverges**.
- If $\rho = 1$, the test is **inconclusive**.

**Formula**:
$$
\rho = \lim_{n \to \infty} n \left( \frac{a_n}{a_{n+1}} - 1 \right)
$$
$$
\begin{cases}
\rho > 1 & \implies \text{Converges}, \\
\rho < 1 & \implies \text{Diverges}, \\
\rho = 1 & \implies \text{Inconclusive}.
\end{cases}
$$

## Summary of Tests

| **Test**                  | **Condition for Convergence**                               | **Condition for Divergence**                     |
|---------------------------|-------------------------------------------------------------|--------------------------------------------------|
| **Nth-Term Test**         | $\lim_{n \to \infty} a_n = 0$                            | $\lim_{n \to \infty} a_n \neq 0$              |
| **Geometric Series**      | $|r| < 1$                                                 | $|r| \geq 1$                                   |
| **p-Series**              | $p > 1$                                                   | $p \leq 1$                                     |
| **Integral Test**         | $\int_{N}^{\infty} f(x) \, dx$ converges                | $\int_{N}^{\infty} f(x) \, dx$ diverges      |
| **Direct Comparison**     | $\sum b_n$ converges                                      | $\sum a_n$ diverges                            |
| **Limit Comparison**      | $\lim \frac{a_n}{b_n} = c \in (0, \infty)$              | Same as $\sum b_n$                             |
| **Alternating Series**    | $\lim b_n = 0$ and $b_n$ decreasing                    | Otherwise inconclusive                           |
| **Ratio Test**            | $\rho < 1$                                                | $\rho > 1$                                     |
| **Root Test**             | $\rho < 1$                                                | $\rho > 1$                                     |
| **Cauchy's Condensation** | Condensed series converges                                  | Condensed series diverges                         |
| **Raabe's Test**          | $\rho > 1$                                                | $\rho < 1$                                     |

### Additional Notes

#### Do Not Apply Tests Sequentially Without Classification
- Avoid the inefficient approach of applying tests in a fixed sequence until one works. Instead, analyze the structure and characteristics of the series to select the most suitable test.

#### Classifying Series
- **Polynomial Terms**: Series involving polynomials are often best approached with the p-Series Test or Comparison Tests.
- **Exponential Terms**: Series with exponential growth or decay typically benefit from the Ratio or Root Tests.
- **Factorials**: The Ratio Test is particularly effective for series containing factorials.
- **Alternating Signs**: Use the Alternating Series Test, and consider absolute convergence by applying other tests to $\lvert a_n \rvert$.

#### Absolute vs. Conditional Convergence
- A series that converges **absolutely** $\left( \sum \lvert a_n \rvert \text{ converges} \right)$ also **converges**.
- A series that converges **conditionally** does so without the convergence of $\sum \lvert a_n \rvert$.

#### When a Test is Inconclusive
- If a chosen test (e.g., Ratio or Root Test) yields an inconclusive result (e.g., $\rho = 1$), consider alternative tests like the Comparison Test or the Integral Test based on the series' form.

#### Simplifying Series
- **Algebraic Manipulation**: Sometimes, rearranging or simplifying the series can make it fit a known test (e.g., factoring out constants, combining like terms).

#### Series with Non-Positive Terms
- For series with negative or alternating terms, consider absolute convergence by applying tests to $\lvert a_n \rvert$. If $\sum \lvert a_n \rvert$ converges, then $\sum a_n$ converges **absolutely**.

#### Practical Tips
- **Identify Dominant Terms**: Focus on the terms that grow the fastest as \(n\) increases to simplify comparisons.
- **Use Highest Power Terms**: When dealing with polynomials in the numerator and denominator, retain only the highest degree terms for comparison.
